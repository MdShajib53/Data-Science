## Subplot  
##### plt.subplot(nrows, ncols, index)

```python
plt.subplot(1, 2, 1) # Ploating 1st row 1st column
plt.plot(x, y, 'r') 

plt.subplot(1, 2, 2) #Ploating 1st row 2nd column
plt.plot(y, x, 'b')
```

## Object Oriented Method

```python
fig = plt.figure()

axes = fig.add_axes([0.1, 0.1, 0.8, 0.8]) #manually adding an axes (plot area) to the fig using normalized figure coordinates.
# fig.add_axes([left, bottom, width, height]) ,,, where Each value is a fraction (0 to 1) of the figure's width and height.
axes.plot(x,y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('This is Title')
```

## Figure Size and DPI

```python
fig, axes = plt.subplots(nrows=2, ncols=1, figsize=(8,4))  # figsize=(8,4) => sets the width = 8 inches, height = 4 inches

axes[0].plot(x,y)

axes[1].plot(y,x) 

plt.tight_layout()
```

## Legend  
- legend() is what makes plot's line labels show up in a nice little box so that viewers know what each line means.

```python
fig = plt.figure() # Creates a blank figure
ax = fig.add_axes([0,0,1,1]) # add an axes => [left, bottom, width, height]

ax.plot(x, x**2, label='X Squared')
ax.plot(x, x**3, label='X Cubed')
ax.legend(loc=0)  #loc=0 → Let matplotlib auto-pick the "best" location where the legend fits without overlapping the plot.
```

## Plot  
#### color, linewidth, alpha value, linestyle, marker, markersize, markerfacecolor, markeredgecolor, markeredgewidth 

```python
fig = plt.figure() 
ax = fig.add_axes([0,0,1,1])

ax.plot(x,y, color='purple')
ax.plot(x,y, color='#00FF00')
```

```python
fig = plt.figure() 
ax = fig.add_axes([0,0,1,1])

ax.plot(x,y, color='red')
ax.plot(x,y, color='#FFFF00', linewidth=5)
```

```python
fig = plt.figure() 
ax = fig.add_axes([0,0,1,1])

ax.plot(x,y, color='red')
ax.plot(x,y, color='#00FF00', linewidth=8, alpha=0.3)
```

```python
fig = plt.figure() 
ax = fig.add_axes([0,0,1,1])

ax.plot(x,y, color='red')
ax.plot(x,y, color='#FFFF00', lw = 5, linestyle='-.') # --
```

```python
fig = plt.figure() 
ax = fig.add_axes([0,0,1,1])

ax.plot(x,y, color='yellow')
ax.plot(x,y, color='#00FF00', lw=5, drawstyle='steps')
```

```python
fig = plt.figure() 

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y,color='purple', lw=3, ls='-', marker='o', markersize=10, markerfacecolor='yellow', markeredgewidth=5, markeredgecolor='green')
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])

ax.plot(x,y,color='purple', lw=2, ls='--')

ax.set_xlim([0,1]) # zooming in and showing only a tiny slice of it, The rest of the plot is hidden (not deleted) — just not shown due to axis limits.
ax.set_ylim([0,2]) 
```

## Histogram

```python
from random import sample 
data = sample(range(1, 1000), 100) #sample() is used to select 100 random numbers from that range(1 to 1000) without repetition.
plt.hist(data)
```

```python
data = [np.random.normal(0, std, 100) for std in range(1,4)]

#rectangular box plot 
plt.boxplot(data, vert=True, patch_artist=True)
```
